```python
from visualize_boundary import visualize_boundary


# SVM Parameters
C = 1000
gamma = 10

# rbf -- Radial basis function
clf = svm.SVC(C=C, kernel='rbf', gamma=gamma)
clf.fit(X, Y)
print ('score:', clf.score(X, Y))

plt.figure()
visualize_boundary(X, Y, clf)
plt.show()
```

```python
def data3_params(X, y):
    C_cands = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 1000, 100]
    gamma_cands = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    max_score = -1
    C_pick = -1
    gamma_pick = -1
    for C in C_cands:
        for gamma in gamma_cands:
            clf = svm.SVC(C=C, kernel='rbf', gamma=gamma)
            clf.fit(X, y)
            score = clf.score(X, Y)
            if score > max_score:
                max_score = score
                C_pick = C
                gamma_pick = gamma

    return C_pick, gamma_pick
```

```python
C, gamma = data3_params(X, Y)
print("C", C)
print("gamma", gamma)
# Train the SVM
clf = svm.SVC(C=C, kernel='rbf', gamma=gamma)
clf.fit(X, Y)
print ('score:', clf.score(X, Y))

plt.figure()
visualize_boundary(X, Y, clf)
plt.show()
```
