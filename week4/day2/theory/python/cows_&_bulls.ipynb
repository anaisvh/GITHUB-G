
- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
def candb():
    a = 0
    number = "1038"
    lista = ["1", "0", "3", "8"]
    while a != "STOP":
        cows = 0
        bulls = 0
        a = input("Enter the number or write STOP:")
        a = a.upper()
        print(a)
        if a == number:
            print("You´ve win")
            break
        if a == 'STOP':
            print("game finish")
        else:
            numero = list(a)
            for i, j in zip(lista, numero):   
                if i == j:
                    cows += 1
                elif j in lista:
                    bulls += 1
            print("cows:", cows, "\nbulls:", bulls)

candb()
```
